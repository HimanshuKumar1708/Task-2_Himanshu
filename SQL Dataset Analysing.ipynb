SQL Work

```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_excel("C:/Users/himan/OneDrive/Desktop/Cleaned_data.xlsx")
```

```python
df.head()
```

```python
conn = sqlite3.connect("ecommerce.db")
```

```python
df.to_sql(
    "sales",
    conn,
    if_exists="replace",
    index=False
)
```

```python
query = """
SELECT Product,
       SUM(TotalPrice) AS Revenue
FROM sales
GROUP BY Product
ORDER BY Revenue DESC;
"""

pd.read_sql_query(query, conn)
```

1 Top Selling Products

```python
query = """
SELECT Product,
       SUM(Quantity) AS Total_Sold
FROM sales
GROUP BY Product
ORDER BY Total_Sold DESC;
"""

result = pd.read_sql_query(query, conn)
result
```

```python
pd.read_sql_query(
    "SELECT * FROM sales LIMIT 5",
    conn
)
```

2 Highest Revenue Products

```python
query = """
SELECT Product,
SUM(TotalPrice) AS Revenue
FROM sales
GROUP BY Product
ORDER BY Revenue DESC;
"""

result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT PaymentMethod,
COUNT(*) AS Orders
FROM sales
GROUP BY PaymentMethod;
"""

result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT strftime('%Y-%m', Date) AS Month,
SUM(TotalPrice) AS Revenue
FROM sales
GROUP BY Month;
"""

result = pd.read_sql_query(query, conn)
result
```

```python
query = """
SELECT CustomerID,
SUM(TotalPrice) AS Spending
FROM sales
GROUP BY CustomerID
ORDER BY Spending DESC
LIMIT 10;
"""

result = pd.read_sql_query(query, conn)
result
```
